```python
using Gadfly
using MultivariateStats
```

```python
N = 100
P = zeros(2, N)
for i = 1:N
    p = rand(2) - 0.5
    p = p / norm(p)
    p[1,:] = p[1,:] * 2.0
    p += .1* (rand() - .5 )
    P[:,i] = p
end

# get rotation
θ = π / 6
R = [cos(θ) -sin(θ); sin(θ) cos(θ)]
# get translation
T = [.2, 2.0]

P_ = zeros(2, N)
for i = 1:N
    P_[:,i] = R * P[:,i] + T + .1* (rand() - .5 )
end
```

```python
plot(layer(x=P[1,:], y=P[2,:], Geom.point, Theme(default_color=color("red"))),
     layer(x=P_[1,:], y=P_[2,:], Geom.point, Theme(default_color=color("green"))),
     Coord.cartesian(fixed=true), Guide.title("Toy example"))
```

## Goal: shift green to align with red

```python
# compute center of mass
c_green = mean(P_, 2)
c_red = mean(P, 2)

# PCA
M_green = fit(PCA, P_)
M_red = fit(PCA, P)

# plot principle components
t = linspace(-3, 3,2)
plot(layer(x=P[1,:], y=P[2,:], Geom.point, Theme(default_color=color("red"))),
     layer(x=P_[1,:], y=P_[2,:], Geom.point, Theme(default_color=color("green"))),
     layer(x= t .* projection(M_green)[1,1]+c_green[1], y = t .* projection(M_green)[2,1]+c_green[2], Geom.line, Theme(default_color=color("green"))),
     layer(x= t .* projection(M_green)[1,2]+c_green[1], y = t .* projection(M_green)[2,2]+c_green[2], Geom.line, Theme(default_color=color("green"))),
      layer(x= t .* projection(M_red)[1,1]+c_red[1], y = t .* projection(M_red)[2,1]+c_red[2], Geom.line, Theme(default_color=color("red"))),
     layer(x= t .* projection(M_red)[1,2]+c_red[1], y = t .* projection(M_red)[2,2]+c_red[2], Geom.line, Theme(default_color=color("red"))),
     Coord.cartesian(fixed=true), Guide.title("PCA's"))
```

```python
# get rotation matrix. want R M_green = M_red
R = transpose(projection(M_green)) * projection(M_red);

# get alignment
P_green_aligned = zeros(2, N)
for i = 1:N
    P_green_aligned[:,i]= R * (P_[:,i] - c_green) + c_red
end
```

```python
plot(layer(x=P[1,:], y=P[2,:], Geom.point, Theme(default_color=color("red"))),
     layer(x=P_green_aligned[1,:], y=P_green_aligned[2,:], Geom.point, Theme(default_color=color("green")))  ,
     Coord.cartesian(fixed=true), Guide.title("Toy example"))
```
